```python
def change(x):
    if "<img" in x:
        x1 = re.findall(r'<p(\s.*?)?>(.*?)</p>', x, flags=re.IGNORECASE | re.MULTILINE | re.DOTALL)
        regex = re.compile('<.*?>')
        x2 = regex.sub('',x1[0][1])
        
        return x2
    else:
        return x
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
x2 = [word_tokenize(i) for i in x1]
```

```python
len(x2)
```

```python
x3 = []

for words in x2:
    clean = []
    for w in words:
        res = re.sub(r'[^\w\s]',"",w)
        if res != "":
            clean.append(res)
    x3.append(clean)
```

```python
len(x3)
```

```python
from nltk.corpus import stopwords
```

```python
x4 = []

for words in x3:
    w=[]
    for word in words:
        if word not in stopwords.words('english'):
            w.append(word)
    x4.append(w)
```

```python
len(x4)
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
pos = PorterStemmer()
```

```python
x5 = []

for words in x4:
    word= []
    for w in words:
        word.append(pos.stem(w))
    x5.append(word)
```

```python
len(x5)
```
